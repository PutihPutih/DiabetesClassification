# Balancing Act

```python
X=df.drop("Diabetes_binary",axis=1)
Y=df["Diabetes_binary"]
```

```python
Y.value_counts()
```

```python
Y.value_counts().plot(kind = 'bar', title = 'Label Distribution')
```

## Balancing Act: Using SMOTE to Tackle Class Imbalance

```python
smote = SMOTE(sampling_strategy='minority')
X_train_resampled, y_train_resampled = smote.fit_resample(X, Y)
```

```python
X_train_resampled.shape, y_train_resampled.shape
```

```python
y_train_resampled.value_counts()
```

```python
y_train_resampled.value_counts().plot(kind = 'bar', title = 'Label Distribution')
```

## Modeling

```python
# Instantiate logistic regression, XGBoost models, and Random Forest
lr = LogisticRegression(max_iter=1500)
xgb = XGBClassifier()
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Use cross_val_predict to generate predicted y values for each fold
y_pred_lr = cross_val_predict(lr, X_train_resampled, y_train_resampled, cv=5)
y_pred_xgb = cross_val_predict(xgb, X_train_resampled, y_train_resampled, cv=5)
y_pred_rf = cross_val_predict(rf, X_train_resampled, y_train_resampled, cv=5)


# Generate confusion matrices for each model and fold
cm_lr = confusion_matrix(y_train_resampled, y_pred_lr)
cm_xgb = confusion_matrix(y_train_resampled, y_pred_xgb)
cm_rf = confusion_matrix(y_train_resampled, y_pred_rf)

# Plot confusion matrices using seaborn heatmap
sns.heatmap(cm_lr, annot=True, cmap='Blues', fmt='g')
plt.title('Logistic Regression Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

sns.heatmap(cm_xgb, annot=True, cmap='Blues', fmt='g')
plt.title('XGBoost Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

sns.heatmap(cm_rf, annot=True, cmap='Blues', fmt='g')
plt.title('XGBoost Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

print(classification_report(y_train_resampled, y_pred_lr))
print(classification_report(y_train_resampled, y_pred_xgb))
print(classification_report(y_train_resampled, y_pred_rf))

```

# Conclusion

```python
# Calculate classification metrics for each model
lr_precision, lr_recall, lr_f1, _ = precision_recall_fscore_support(y_train_resampled, y_pred_lr, average='binary')
xgb_precision, xgb_recall, xgb_f1, _ = precision_recall_fscore_support(y_train_resampled, y_pred_xgb, average='binary')
rf_precision, rf_recall, rf_f1, _ = precision_recall_fscore_support(y_train_resampled, y_pred_rf, average='binary')

# Create a bar chart for comparison
labels = ['Logistic Regression', 'XGBoost', 'Random Forest']
precision_scores = [lr_precision, xgb_precision, rf_precision]
recall_scores = [lr_recall, xgb_recall, rf_recall]
f1_scores = [lr_f1, xgb_f1, rf_f1]

x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots(figsize=(15,7))
rects1 = ax.bar(x - width, precision_scores, width, label='Precision')
rects2 = ax.bar(x, recall_scores, width, label='Recall')
rects3 = ax.bar(x + width, f1_scores, width, label='F1-score')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Model Comparison')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
fig.tight_layout()

plt.show()
```
